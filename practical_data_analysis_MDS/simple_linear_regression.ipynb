## [Part-2](https://www.youtube.com/watch?v=H3-v21v92N0&list=PLD_eiqVVLZDj1onVQPuREzf90xRT2WIdR&index=41)

```python
X = df[['Schooling']]
Y = df[['life_expect']]
X.head()
```

```python
Y.head()
```

```python
from sklearn.linear_model import LinearRegression
slr = LinearRegression()
slr
```

```python
slr.fit(X,Y)
```

```python
Yhat = slr.predict(X)
Yhat[0:5]
```

```python
df.insert(loc=4, column='predict_life_expect', value=Yhat,allow_duplicates=True)
df.head()
```

```python
df['Schooling'].min(), df['Schooling'].max() 
# predict values must be between this two values
```

```python
import numpy as np
slr.predict(np.array([15]).reshape(-1,1))
```

```python
slr.predict(np.array([5]).reshape(-1,1))
```

```python
slr.coef_       # slope value (m)
```

```python
slr.intercept_  # y intercept value (b)
```
