```python
# data frame ထဲက null ဖြစ်နေတာတွေကိုပဲရွေးထုတ်တာ
data[data.isnull().any(axis=1)]
```

```python
data.count()
```

```python
data.dropna(axis=1)
```

### Filling NA
average value assign လုပ်နည်း

```python
avg_weight = round(data['weight'].mean(),2)
avg_weight
```

```python
data['weight'] = data['weight'].fillna(avg_weight)
data.head(20)
```

```python
data.count()
```

### Formatting

```python
data.birth_place.unique()
```

```python
data['birth_place'] = data['birth_place'].str.replace('NY', 'New York')
data.head()
```

```python
data['birth_place'] = data['birth_place'].str.replace('LA', 'Los Angeles')
data.head()
```

### Converting

```python
data['weight'] = data['weight'] * 2.205
data.head()
```
